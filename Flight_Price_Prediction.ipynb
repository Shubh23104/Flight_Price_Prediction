## ***Feature Engineering Process***

```python
# handling Date time Feature
final_df['Date'] = [feature for feature in final_df['Date_of_Journey'].str.split('/').str[0]]
final_df['Month'] = [feature for feature in final_df['Date_of_Journey'].str.split('/').str[1]]
final_df['Year'] = [feature for feature in final_df['Date_of_Journey'].str.split('/').str[2]]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df = final_df.drop(['Date_of_Journey'],axis=1)
```

```python
final_df.head(3)
```

```python
# Handling Arrival time Category
final_df['Arrival_Time'] = [feature for feature in final_df['Arrival_Time'].str.split(' ').str[0]]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_hour'] = [feature for feature in final_df['Arrival_Time'].str.split(':').str[0]]
final_df['Arrival_minute'] = [feature for feature in final_df['Arrival_Time'].str.split(':').str[1]]
```

```python
final_df.head()
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_minute'] = final_df['Arrival_minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop(['Arrival_Time'],axis=1,inplace=True)
```

```python
final_df.head(3)
```

```python
# Handling Depature time Category
final_df['Dep_Time'] = [feature for feature in final_df['Dep_Time'].str.split(' ').str[0]]
```

```python
final_df.head(1)
```

```python
final_df['Dept_hour'] = [feature for feature in final_df['Dep_Time'].str.split(':').str[0]]
final_df['Dept_minute'] = [feature for feature in final_df['Dep_Time'].str.split(':').str[1]]
final_df['Dept_hour'] = final_df['Dept_hour'].astype(int)
final_df['Dept_minute'] = final_df['Dept_minute'].astype(int)
```

```python
final_df.head(3)
```

```python
final_df.drop(['Dep_Time'],axis=1,inplace=True)
```

```python
final_df.head(3)
```

```python
final_df.info()
```

```python
final_df.head(2)
```

```python
final_df.loc[final_df['Total_Stops'].isnull()]
```

```python
# Handling Route and Total Stops Category
# But here I can See that route is similar to total stops so i can taken information from total stops and i can not require route feature
# but i can not drop it because stops are influence the price when number of stops are increse the price are incresed
final_df['Total_Stops'].unique()
```

```python
final_df.loc[final_df['Total_Stops'] == 'nan']
```

```python
final_df.loc[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```

```python
final_df.head(3)
```

```python
final_df.drop(['Route'],axis=1,inplace=True)
```

```python
final_df.head(3)
```

```python
final_df.info()
```

```python
# handling Additional info Category
final_df['Additional_Info'].unique()
```

```python
final_df.replace({"Additional_Info":['No Info']}, {"Additional_Info":['No info']}, inplace=True)
```

```python
final_df['Additional_Info'].unique()
```

```python
# Handling Duration Category
final_df['d_hour'] = [feature for feature in final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]]
```

```python
final_df.head()
```

```python
final_df['d_minute'] = [feature for feature in final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]]
```

```python
final_df.head()
```

```python
final_df['d_minute'] = final_df['d_minute'].fillna(0)
final_df['d_minute'] = final_df['d_minute'].astype(int)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['d_hour'].unique()
```

```python
final_df.replace({"d_hour":['5m']}, {"d_hour":['0']}, inplace=True)
```

```python
final_df['d_hour'] = final_df['d_hour'].astype(int)
```

```python
final_df.info()
```

```python
final_df.loc[final_df['d_hour'] == 0]
```

```python
final_df['Duration'] = (final_df['d_hour'] * 60) + final_df['d_minute']
```

```python
final_df.head()
```

```python
final_df.loc[final_df['Duration'] == 0]
```

```python
final_df.replace({"Duration":[0]}, {"Duration":[5]}, inplace=True)
```

```python
final_df.loc[final_df['Duration'] == 5]
```

```python
final_df.info()
```

```python
final_df.drop(['d_hour'],axis=1,inplace=True)
```

```python
final_df.drop(['d_minute'],axis=1,inplace=True)
```

```python
final_df.info()
```
